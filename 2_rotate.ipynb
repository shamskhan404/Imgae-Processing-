```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('happy2.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Image rotation parameter
center = (img_rgb.shape[1] // 2, img_rgb.shape[0]//2)
angle = 30
scale = 1
```

```python
# getRotationMatrix2D creates a matrix needed for transformation.
rotation_matrix = cv2.getRotationMatrix2D(center, angle,scale)
```

```python
# We want matrix for rotation w.r.t center to 30 degree without scaling.
rotated_img = cv2.warpAffine(img_rgb, rotation_matrix,(img.shape[1], img.shape[0]))
```

```python
fig , axs = plt.subplots(1,2, figsize=(7,4))

axs[0].imshow(img_rgb)
axs[0].set_title('Org Img')

axs[1].imshow(rotated_img)
axs[1].set_title('Rotated Img')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
