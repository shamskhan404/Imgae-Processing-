```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('happy2.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
width = img_rgb.shape[1]
height = img_rgb.shape[0]

tx = 100
ty = 70

# Translation matrix
import numpy as np
translation_matrix = np.array([[1,0, tx], [0,1, ty]],dtype = np.float32)
```

```python
# warpAffine does appropriate shifting given the Translation matrix.

translated_img = cv2.warpAffine(img_rgb,translation_matrix,(width,height))
```

```python
fig,axs = plt.subplots(1,2,figsize=(7,4))
axs[0].imshow(img_rgb)
axs[0].set_title('Org Img')

axs[1].imshow(translated_img)
axs[1].set_title('Translation Img')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
