> **The process of image edge detection involves detecting sharp edges in the image**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('happy2.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Apply canny edge detection 
edges = cv2.Canny( img_rgb , threshold1=100,threshold2=700)
```

> **Threshold 1** *(Lower Threshold): This is the minimum intensity gradient value. Pixels with gradient values below this threshold are automatically discarded and not considered as edges.*

> **Threshold 2** *(Upper Threshold): This is the maximum intensity gradient value. Pixels with gradient values above this threshold are considered strong edges and retained.*

```python
fig ,axs = plt.subplots(1,2, figsize=(7,4))

axs[0].imshow(img_rgb)
axs[0].set_title('Original Img')

axs[1].imshow(edges)
axs[1].set_title('Img edges')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout
plt.show()
```
