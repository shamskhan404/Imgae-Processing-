> **Image shearing is a geometric transformation that skews an image along one or both axes i.e x or y axis.**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('happy2.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

width = img_rgb.shape[1]
height = img_rgb.shape[0]

# Define the Shearing factor
shear_x = -0.15
shear_y = 0
```

```python
# Define the Transformation matrix for shearing
transformation_matrix = np.array([[1, shear_x, 0],[0,1,shear_y]],dtype = np.float32)
```

```python
# Apply shearing
sheared_img = cv2.warpAffine(img_rgb, transformation_matrix,(width,height))
```

```python
fig,axs = plt.subplots(1,2,figsize=(7,4))
axs[0].imshow(img_rgb)
axs[0].set_title('Org Img')

axs[1].imshow(sheared_img)
axs[1].set_title('Sheared Img')

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
