```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('happy2.jpg')

# Convert BGR image to RGB
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define the scale factor
# Increase the size by 3 times
scl_1 = 3.0  
# Decrease the size by 3 times
scl_2 = 1/3.0

# Get the original image dimensions
height, width = img_rgb.shape[:2]
```

```python
new_height = int(height * scl_1)
new_width = int(width * scl_1)

# Resize the image
zoomed_img = cv2.resize(src =img_rgb, 
                          dsize=(new_width, new_height), 
                          interpolation=cv2.INTER_CUBIC)
```

```python
new_height1 = int(height * scl_2)
new_width1 = int(width * scl_2)

# Scaled image
scaled_img = cv2.resize(src= img_rgb, 
                          dsize =(new_width1, new_height1), 
                          interpolation=cv2.INTER_AREA)
```

```python
# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img_rgb)
axs[0].set_title('Original Image Shape:'+str(img_rgb.shape))

# Plot the Zoomed Image
axs[1].imshow(zoomed_img)
axs[1].set_title('Zoomed Image Shape:'+str(zoomed_img.shape))

# Plot the Scaled Image
axs[2].imshow(scaled_img)
axs[2].set_title('Scaled Image Shape:'+str(scaled_img.shape))

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])
# Display the subplots  
plt.tight_layout()
plt.show()
```
